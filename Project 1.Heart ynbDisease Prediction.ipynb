importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing


```python
# loading the csv data to a Pandas Dtaframe
heart_data=pd.read_csv("/content/heart.csv")
```

SPLITTING THE FEATURE AND TARGET


```python
X=heart_data.drop(columns="target", axis=1)
Y=heart_data["target"]
```

```python
print(X)
```

```python
print(Y)
```

Splitting Data into Training data & Test Data

```python
X_train , X_test, Y_train, Y_test =train_test_split( X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X_train.shape, X_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit( X_train, Y_train)
```

Model Evalution
Accuracy Score

```python
# accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on tTraining data:",training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on testing data:",test_data_accuracy)
```

```python
import gradio as gr
import pickle
import numpy as np

# Load the saved model

model = pickle.load(open("heart_disease_model.pkl", "rb"))


# Define prediction function
def predict_heart_disease(age, sex, cp, trestbps, cholesterol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal):
    input_data = np.array([age, sex, cp, trestbps, cholesterol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal], dtype=float).reshape(1, -1)
    prediction = model.predict(input_data)[0]

    if prediction == 1:
        return "🔴 *Heart Disease Detected!*"
    else:
        return "🟢 *No Heart Disease!*"

# Custom CSS for styling
custom_css = """
body {background-color: #f5f7fa;}
h1 {color: #007bff; text-align: center;}
.gradio-container {border: 2px solid #007bff; padding: 20px; background: white; border-radius: 10px;}
"""

# Create Gradio interface
interface = gr.Interface(
    fn=predict_heart_disease,
    inputs=["number"] * 13,  # 13 numeric inputs
    outputs="text",  # Display prediction as text
    title="💙 Heart Disease Prediction",
    description="Enter values like *age, cholesterol, blood pressure, etc.* to predict heart disease.",
    theme="default",
    css=custom_css
)

# Launch interface
interface.launch()
```
